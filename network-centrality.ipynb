```python
#node-degree distribution
import pylab as plt

G = nx.read_gml('C:\\Users\\hp\\power.gml','id')
idegrees = dict(G.degree()) # dictionary node:degree
ivalues = sorted(set(idegrees.values()))
ihist = [list(idegrees.values()).count(x) for x in ivalues]

plt.figure() 
plt.grid(True)
plt.plot(ivalues, ihist, 'ro-') 
plt.legend(['degree'])
plt.xlabel('Degree')
plt.ylabel('Number of nodes')
plt.title('Node-Degree Distribution')
plt.xlim([0, 2*4**2])
plt.savefig('./DegreeDistrubtion.pdf')
plt.show()
plt.close()
```

```python
#node-degree distribution
import pylab as plt

G = nx.read_gml('C:\\Users\\hp\\power.gml','id')
idegrees = dict(G.degree()) # dictionary node:degree
ivalues = sorted(set(idegrees.values()))
ihist = [list(idegrees.values()).count(x) for x in ivalues]

plt.figure() 
plt.grid(True)
plt.loglog(ivalues, ihist, 'ro-') 
plt.legend(['degree'])
plt.xlabel('Degree')
plt.ylabel('Number of nodes')
plt.title('Node-Degree Distribution')
plt.xlim([0, 2*10**2])
plt.savefig('./DegreeDistrubtion.pdf')
plt.show()
plt.close()
```
